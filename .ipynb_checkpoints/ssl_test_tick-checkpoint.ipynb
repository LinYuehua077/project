```python
import numpy as np
import pandas as pd
```

```python
# 数据导入
total_df = pd.read_csv("total_df.csv")
```

```python
total_df.head()
```

```python
total_df.shape
```

```python
total_df['date'] = pd.to_datetime(total_df['date'])                             # 将日期设为索引
total_df = total_df.sort_values(['date', 'code']).reset_index(drop=True)        # 按[日期,股票代码]排序，设置二级索引
total_df = total_df.join(total_df.set_index('date'), on='date')                 # 按日期对齐
```
